```python
# Flight Search Function

# Advanced Flight Search Function

def search_flights(
    from_city,
    to_city,
    travel_class,
    departure_time=None,
    max_price=None
):

    filtered_df = flights_df[

        (flights_df["from_city"].str.lower() == from_city.lower()) &

        (flights_df["to_city"].str.lower() == to_city.lower()) &

        (flights_df["class"].str.lower() == travel_class.lower())

    ]

    # Time filtering

    if departure_time:

        filtered_df = filtered_df[

            filtered_df["departure_time"].str.lower()
            ==
            departure_time.lower()

        ]

    # Budget filtering

    if max_price:

        filtered_df = filtered_df[
            filtered_df["price"] <= max_price
        ]

    # Cheapest flights first

    filtered_df = filtered_df.sort_values(
        by="price"
    )

    return filtered_df.head(5)
```

```python
# Create Ticket Booking Function
# Ticket Booking Function

import random

def book_ticket(

    passenger_name,
    from_city,
    to_city,
    airline,
    travel_class,
    departure_time,
    travel_date,
    seat_preference,
    price

):

    ticket_id = f"AI-{random.randint(10000,99999)}"

    # Seat Logic
    if seat_preference.lower() == "window":
        seat_letter = random.choice(["A", "F"])

    elif seat_preference.lower() == "aisle":
        seat_letter = random.choice(["C", "D"])

    else:
        seat_letter = random.choice(["B", "E"])

    seat_number = f"{random.randint(1,30)}{seat_letter}"

    ticket = {

        "ticket_id": ticket_id,

        "passenger_name": passenger_name,

        "from_city": from_city,

        "to_city": to_city,

        "airline": airline,

        "travel_class": travel_class,

        "departure_time": departure_time,

        "travel_date": travel_date,

        "seat_preference": seat_preference,

        "seat_number": seat_number,

        "price": price,

        "booking_status": "CONFIRMED"

    }

    return ticket
```
